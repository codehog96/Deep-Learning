### Plotting a Simple Neural Network

```python
import matplotlib.pyplot as plt
```

```python
def plot_neural_network():
    # Define the positions of the layers
    input_layer = [-1, 0, 1]  # Three nodes
    hidden_layer = [-1.5, 0, 1.5]  # Three nodes
    output_layer = [0]  # One node

    # Plot nodes
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.scatter([0]*len(input_layer), input_layer, s=500, color="skyblue", label="Input Layer")
    ax.scatter([1]*len(hidden_layer), hidden_layer, s=500, color="orange", label="Hidden Layer")
    ax.scatter([2]*len(output_layer), output_layer, s=500, color="lightgreen", label="Output Layer")
    
    # Plot connections (lines between nodes)
    for i in input_layer:
        for j in hidden_layer:
            ax.plot([0, 1], [i, j], 'k-', lw=1)
    for j in hidden_layer:
        for k in output_layer:
            ax.plot([1, 2], [j, k], 'k-', lw=1)

    # Add labels
    for i, pos in enumerate(input_layer):
        ax.text(-0.1, pos, f"I{i+1}", fontsize=12, ha='right')
    for i, pos in enumerate(hidden_layer):
        ax.text(1.1, pos, f"H{i+1}", fontsize=12, ha='left')
    ax.text(2.1, 0, "O1", fontsize=12, ha='left')

    # Set axis limits and title
    ax.set_xlim(-0.5, 2.5)
    ax.set_ylim(-2, 2)
    ax.axis('off')  # Hide axes
    ax.set_title("Simple Neural Network", fontsize=16)
    plt.legend(loc="upper right")
    plt.show()

# Call the function to plot
plot_neural_network()
```

### Explanation:
	1.	Layers: The positions of the input, hidden, and output layers are defined.
	2.	Nodes: Scatter plots are used to represent the neurons in each layer.
	3.	Connections: Lines (ax.plot) are drawn to connect nodes between layers.
	4.	Labels: Nodes are labeled for better understanding.
	5.	Visualization: The plot is displayed with Matplotlib.
